Questão 2 Clusten Usando k-mens

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Vetores tridimensionais
X = np.array([
    [-5.82, -4.58, -1.97],
    [-4.68, 2.16, 3.71],
    [3.36, -3.19, 1.09],
    [7.72, 0.88, 1.80],
    [-7.64, 3.06, 3.50],
    [-6.87, 0.57, -5.45],
    [4.47, -2.62, 5.76],
    [7.73, -2.01, 5.18],
    [-7.71, 3.34, -6.33],
    [-5.91, -0.49, -5.68],
    [2.18, 3.81, 5.82],
    [6.72, -0.93, -3.04],
    [-5.25, -0.26, 0.56],
    [-6.94, -1.22, 1.13],
    [7.09, 0.20, 2.25],
    [6.81, 3.17, -4.15],
    [-4.19, 4.24, 4.04],
    [-5.38, -1.74, 1.43],
    [5.08, 3.30, 5.33],
    [7.27, 0.93, -2.78]
])

```

```python
# Número de clusters desejado
num_clusters = 3

# Cria o objeto KMeans e realiza a clusterização
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)

# Obtem os rótulos dos clusters para cada vetor
labels = kmeans.labels_

# Obtem as coordenadas dos centroides de cada cluster
centroids = kmeans.cluster_centers_


```

```python
# Plotando os gráficos 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotando os vetores
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels, cmap='viridis', s=50)

# Plotando os centroides
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], c='red', marker='x', s=100, label='Centroides')

# Configurações dos rótulos dos eixos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Exibindo a legenda
ax.legend()

# Exibindo o gráfico
plt.show()
```

b) Centroides Inicialmente definidos

```python
# Coordenadas dos centroides iniciais
centroides_iniciais = np.array([[0, 0, 0], [1, 1, 1], [-1, 0, 2]])

# Número de clusters desejado
num_clusters = 3

# Cria o objeto KMeans e realiza a clusterização
kmeans = KMeans(n_clusters=num_clusters, init=centroides_iniciais)
kmeans.fit(X)

# Obtem os rótulos dos clusters para cada vetor
labels = kmeans.labels_

# Obtem as coordenadas dos centroides de cada cluster
centroids = kmeans.cluster_centers_
```

```python
# Plotando os gráficos 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotando os vetores
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels, cmap='viridis', s=50)

# Plotando os centroides
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], c='red', marker='x', s=100, label='Centroides')

# Configurações dos rótulos dos eixos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Exibindo a legenda
ax.legend()

# Exibindo o gráfico
plt.show()
```
